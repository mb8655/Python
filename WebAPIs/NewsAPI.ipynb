```python
# prompt: from the above request users to enter the keyword

import pandas as pd
import re
from bs4 import BeautifulSoup
import requests
import numpy as np
# set the environment
#sign up and obtain NewsAPI.org api key

news_api_key = '88c0f8d156134bf38a34493f9958fca7'
endpoint = 'https://newsapi.org/v2/top-headlines'


def get_news_by_keyword(keyword):
  parameters = {
      'q': keyword,  # Add the keyword to the query parameter
      'apiKey': news_api_key,
      'pageSize': 100
  }

  response = requests.get(endpoint, params=parameters)
  data = response.json()
  return data['articles'] if data['status'] == 'ok' else []


# Example usage:
keyword = input("Enter a keyword to search for news: ")  # Get keyword from user
articles = get_news_by_keyword(keyword)
if articles:
    print(f"Articles related to '{keyword}':")
    for article in articles:
        print(article['title'])
else:
    print(f"No articles found for '{keyword}'.")
```
